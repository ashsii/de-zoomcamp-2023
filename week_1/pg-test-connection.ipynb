````python
query = """
SELECT to_date(lpep_pickup_datetime::TEXT,'YYYY-MM-DD') ,* 
FROM green_taxi_data 
where to_date(lpep_pickup_datetime::TEXT,'YYYY-MM-DD') = '2019-01-15'
and to_date(lpep_dropoff_datetime::TEXT,'YYYY-MM-DD') = '2019-01-15'
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT * 
FROM green_taxi_data 
order by trip_distance desc
limit 10
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT * 
FROM green_taxi_data 
where passenger_count = 2
and to_date(lpep_pickup_datetime::TEXT,'YYYY-MM-DD') = '2019-01-01'
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT * 
FROM green_taxi_data 
where passenger_count = 3
and to_date(lpep_pickup_datetime::TEXT,'YYYY-MM-DD') = '2019-01-01'
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT * 
FROM zones
where "Zone" = 'Astoria'
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT * 
FROM green_taxi_data g
left join zones z on g."DOLocationID" = z."LocationID"
where "PULocationID" = 7 
order by tip_amount desc
limit 10
"""

pd.read_sql(query, con=engine)
````
